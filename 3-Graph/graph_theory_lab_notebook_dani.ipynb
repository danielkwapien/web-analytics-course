- *Note that you cannot use any of the functions that automatically creates a random graph from NetworkX library.*

````python
#Solution Milestone 1 question a) 

# Parameters
num_nodes = 10  # Number of nodes
p = 0.4         # Probability of edge existence between pairs

# Initialize an empty undirected graph
G = nx.Graph()

# Add nodes to the graph
G.add_nodes_from(range(num_nodes))

# Iterate through all unique pairs of nodes to add edges based on probability p
for i in range(num_nodes):
    for j in range(i + 1, num_nodes):  # Only consider pairs once (i, j) where i < j for an undirected graph
        if random.random() < p:  # Add an edge with probability p
            G.add_edge(i, j)

pos = nx.spring_layout(G)  # Define layout for visualization
nx.draw_networkx(G, pos, with_labels=True, node_color="lightblue", edge_color="gray", font_weight="bold")
plt.title("Undirected Random Graph (Gilbert Model) with 10 Nodes, p=0.4")
plt.show()
````

````python
#Solution Milestone 1 question b)

# Parameters
num_nodes = 10  # Number of nodes
p = 0.4         # Probability of directed edge existence between pairs

# Initialize an empty directed graph
G = nx.DiGraph()

# Add nodes to the graph
G.add_nodes_from(range(num_nodes))

# Iterate through all ordered pairs of nodes to add directed edges based on probability p
for i in range(num_nodes):
    for j in range(num_nodes):
        if i != j:  # Exclude self-loops
            if random.random() < p:  # Add a directed edge with probability p
                G.add_edge(i, j)

pos = nx.spring_layout(G)  # Define layout for visualization
nx.draw_networkx(G, pos, with_labels=True, node_color="lightcoral", edge_color="gray", font_weight="bold", arrows=True)
plt.title("Directed Random Graph (Gilbert Model) with 10 Nodes, p=0.4")
plt.show()
````

````python
#Solution Milestone 2, question c) 
# Step 2: Manually compute Closeness Centrality with NetworkX normalization

def compute_closeness_centrality(graph, node):
    # Reverse graph if directed for inward reachability calculation
    if graph.is_directed():
        graph = graph.reverse()  # create a reversed graph view

    # Use single_source_shortest_path_length to get shortest paths
    path_length = nx.single_source_shortest_path_length

    # Handle node input (for compatibility, we'll assume node is provided)
    nodes = [node] if node is not None else graph.nodes
    closeness_dict = {}
    
    for n in nodes:
        sp = path_length(graph, n)  # Get shortest paths from n
        totsp = sum(sp.values())    # Sum of path lengths
        len_G = len(graph)          # Total number of nodes in the graph
        _closeness_centrality = 0.0
        
        # Calculate closeness if there are reachable nodes and more than 1 node in graph
        if totsp > 0.0 and len_G > 1:
            _closeness_centrality = (len(sp) - 1.0) / totsp
            # Normalize based on connected part
            s = (len(sp) - 1.0) / (len_G - 1)
            _closeness_centrality *= s
        closeness_dict[n] = _closeness_centrality
    return closeness_dict[node]


# Calculate closeness centrality for each node manually
closeness_centrality_manual = {node: compute_closeness_centrality(G, node) for node in G.nodes}

# Step 3: Calculate using NetworkX function for comparison
closeness_centrality_nx = nx.closeness_centrality(G)

# Print results
print("Manual Closeness Centrality:", closeness_centrality_manual)
print("NetworkX Closeness Centrality:", closeness_centrality_nx)

# Verify if manual calculations match NetworkX results
closeness_match = all(abs(closeness_centrality_manual[node] - closeness_centrality_nx[node]) < 1e-5 for node in G.nodes)
print("\nCloseness Centrality Match:", closeness_match)
````

````python
from collections import deque
#Solution Milestone 2, question d)
def compute_betweenness_centrality(graph):
    betweenness = {node: 0.0 for node in graph.nodes}
    for s in graph.nodes:
        # Initialize data structures
        stack = []
        pred = {w: [] for w in graph.nodes}  # List of predecessors
        sigma = dict.fromkeys(graph.nodes, 0)  # Shortest path counts
        sigma[s] = 1
        dist = dict.fromkeys(graph.nodes, -1)  # Distance from source
        dist[s] = 0
        queue = deque([s])

        # Step 2.1: BFS to calculate shortest paths
        while queue:
            v = queue.popleft()
            stack.append(v)
            for w in graph.successors(v):
                # Path discovery
                if dist[w] < 0:
                    queue.append(w)
                    dist[w] = dist[v] + 1
                # Path counting
                if dist[w] == dist[v] + 1:
                    sigma[w] += sigma[v]
                    pred[w].append(v)

        # Step 2.2: Accumulation of betweenness
        delta = dict.fromkeys(graph.nodes, 0)  # Dependency
        while stack:
            w = stack.pop()
            for v in pred[w]:
                delta[v] += (sigma[v] / sigma[w]) * (1 + delta[w])
            if w != s:
                betweenness[w] += delta[w]

    # Normalize for directed graph
    for node in betweenness:
        betweenness[node] /= (len(graph) - 1) * (len(graph) - 2)
    
    return betweenness

# Calculate betweenness centrality manually
betweenness_centrality_manual = compute_betweenness_centrality(G)

# Step 3: Calculate using NetworkX function for comparison
betweenness_centrality_nx = nx.betweenness_centrality(G, normalized=True)

# Print results
print("Manual Betweenness Centrality:", betweenness_centrality_manual)
print("NetworkX Betweenness Centrality:", betweenness_centrality_nx)

# Verify if manual calculations match NetworkX results
betweenness_match = all(abs(betweenness_centrality_manual[node] - betweenness_centrality_nx[node]) < 1e-5 for node in G.nodes)
print("\nBetweenness Centrality Match:", betweenness_match)
````

````python
#Solution Milestone 2, question e) 
damping_factor = 0.85
tolerance = 1e-5

def compute_pagerank(graph, damping_factor=0.85, tolerance=1e-5):
    N = len(graph)
    pagerank = {node: 1 / N for node in graph.nodes}  # Initialize PageRank values
    new_pagerank = pagerank.copy()
    convergence = False
    
    while not convergence:
        # Store previous PageRank sum for convergence check
        total_difference = 0
        for node in graph.nodes:
            rank_sum = sum(pagerank[neighbor] / len(list(graph.successors(neighbor))) 
                           for neighbor in graph.predecessors(node))
            new_pagerank[node] = (1 - damping_factor) / N + damping_factor * rank_sum
        
        # Calculate the sum of differences for convergence
        total_difference = sum(abs(new_pagerank[node] - pagerank[node]) for node in graph.nodes)
        pagerank, new_pagerank = new_pagerank, pagerank  # Update for next iteration
        
        # Check if the change is below tolerance
        convergence = total_difference < tolerance
    
    return pagerank

# Calculate PageRank manually
pagerank_manual = compute_pagerank(G, damping_factor, tolerance)

# Step 3: Calculate using NetworkX function for comparison
pagerank_nx = nx.pagerank(G, alpha=damping_factor, tol=tolerance)

# Print results
print("Manual PageRank:", pagerank_manual)
print("NetworkX PageRank:", pagerank_nx)

# Verify if manual calculations match NetworkX results
pagerank_match = all(abs(pagerank_manual[node] - pagerank_nx[node]) < tolerance for node in G.nodes)
print("\nPageRank Match:", pagerank_match)
````

a) Write a code that  implements the k-core decomposition of a graph. The code should provide the k-core and the layer for each node in the graph. *Note the code cannot use the function defined for this purpose in the NetworkX library*

Print the results of your solution along with the results obtained with the functions of NetworkX library to compute the k-core (```core_number```) and layer (```onion_layers```). **Note tha both, your solution and NetworkX functions should provide the same results**


````python
#Solution Milestone 3 

import networkx as nx
import random

# Parameters for the undirected random graph
num_nodes = 10
p = 0.4

# Step 1: Create an undirected random graph
G = nx.Graph()
G.add_nodes_from(range(num_nodes))

for i in range(num_nodes):
    for j in range(i + 1, num_nodes):
        if random.random() < p:
            G.add_edge(i, j)

# Step 2: Manually compute k-core and onion layers

def k_core_and_onion_layers(graph):
    # Initialize dictionaries to store k-core and onion layer results
    core = {}
    onion_layer = {}
    
    # Get adjacency and degree data
    neighbors = {v: list(graph.neighbors(v)) for v in graph}
    degrees = dict(graph.degree())
    
    # Set initial layer and core level
    current_core = 1
    current_layer = 1
    
    # Handle isolated nodes (if any)
    isolated_nodes = list(nx.isolates(graph))
    if isolated_nodes:
        for node in isolated_nodes:
            onion_layer[node] = current_layer
            core[node] = 0
            degrees.pop(node)
        current_layer += 1
    
    # Process remaining nodes for k-core and onion layers
    while degrees:
        nodes = sorted(degrees, key=degrees.get)
        
        # Update current core level based on the minimum degree
        min_degree = degrees[nodes[0]]
        if min_degree > current_core:
            current_core = min_degree
        
        # Identify nodes in the current layer
        this_layer = []
        for node in nodes:
            if degrees[node] > current_core:
                break
            this_layer.append(node)
        
        # Assign core and layer values, then update neighbors
        for node in this_layer:
            onion_layer[node] = current_layer
            core[node] = current_core
            for neighbor in neighbors[node]:
                neighbors[neighbor].remove(node)
                degrees[neighbor] -= 1
            degrees.pop(node)
        
        # Move to the next layer
        current_layer += 1
    
    return core, onion_layer

# Calculate k-core and layers manually
core_manual, layer_manual = k_core_and_onion_layers(G)

# Step 3: Calculate using NetworkX functions for comparison
core_nx = nx.core_number(G)
try:
    layer_nx = nx.onion_layers(G)
except AttributeError:
    layer_nx = "NetworkX version does not support onion_layers function"

# Print results
print("Manual k-core decomposition:", core_manual)
print("NetworkX k-core decomposition:", core_nx)
print("Manual onion layers:", layer_manual)
print("NetworkX onion layers:", layer_nx)

# Verify if manual calculations match NetworkX results
core_match = core_manual == core_nx
layer_match = layer_manual == layer_nx if layer_nx != "NetworkX version does not support onion_layers function" else "Not Applicable"
print("\nK-core Match:", core_match)
print("Layer Match:", layer_match)
````

````python
#Solution Milestone 4
import time
from networkx.algorithms.community import modularity, label_propagation_communities, greedy_modularity_communities
from community import community_louvain

# Predefined graphs
graphs = {
    "karate_club_graph": nx.karate_club_graph(),
    "davis_southern_women_graph": nx.davis_southern_women_graph(),
    "florentine_families_graph": nx.florentine_families_graph(),
    "les_miserables_graph": nx.les_miserables_graph()
}

# Step 1: Function to calculate modularity manually
def compute_modularity(graph, communities):
    m = graph.number_of_edges()
    degrees = dict(graph.degree())
    Q = 0.0
    for community in communities:
        for u in community:
            for v in community:
                if u != v:
                    A_uv = 1 if graph.has_edge(u, v) else 0
                    Q += (A_uv - (degrees[u] * degrees[v]) / (2 * m))
    return Q / (2 * m)

# Step 2: Apply community detection algorithms and calculate modularity and execution time

results = {}

for graph_name, graph in graphs.items():
    results[graph_name] = {}
    
    # 1. Label Propagation
    start_time = time.time()
    communities = list(label_propagation_communities(graph))
    label_propagation_modularity = compute_modularity(graph, communities)
    label_propagation_time = time.time() - start_time
    results[graph_name]["Label Propagation"] = {
        "modularity": label_propagation_modularity,
        "time": label_propagation_time
    }
    
    # 2. Greedy Modularity
    start_time = time.time()
    communities = list(greedy_modularity_communities(graph))
    greedy_modularity_modularity = compute_modularity(graph, communities)
    greedy_modularity_time = time.time() - start_time
    results[graph_name]["Greedy Modularity"] = {
        "modularity": greedy_modularity_modularity,
        "time": greedy_modularity_time
    }
    
    # 3. Louvain Method
    start_time = time.time()
    partition = community_louvain.best_partition(graph)
    louvain_communities = {}
    for node, community_id in partition.items():
        if community_id not in louvain_communities:
            louvain_communities[community_id] = []
        louvain_communities[community_id].append(node)
    louvain_communities = list(louvain_communities.values())
    louvain_modularity = compute_modularity(graph, louvain_communities)
    louvain_time = time.time() - start_time
    results[graph_name]["Louvain Method"] = {
        "modularity": louvain_modularity,
        "time": louvain_time
    }

# Step 3: Print the results
for graph_name, metrics in results.items():
    print(f"Results for {graph_name}:")
    for algo_name, result in metrics.items():
        print(f"  {algo_name}:")
        print(f"    NX Modularity: {nx.com}")
        print(f"    Modularity: {result['modularity']}")
        print(f"    Execution Time: {result['time']} seconds")

# Optional validation of modularity calculation with NetworkX
print("\nValidation with NetworkX modularity function:")
for graph_name, graph in graphs.items():
    communities = list(greedy_modularity_communities(graph))  # Using Greedy Modularity for example
    nx_modularity = modularity(graph, communities)
    manual_modularity = compute_modularity(graph, communities)
    print(f"{graph_name} - NetworkX Modularity: {nx_modularity}, Manual Modularity: {manual_modularity}")
````
