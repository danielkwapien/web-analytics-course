a) Write the code to construct from scratch an **unidirected** random graph with 10 nodes using the Gilbert model, i.e., the probability of an edge existing between a pair of nodes is p. For this exercise use p = 0.4.

- *Note that you cannot use any of the functions that automatically creates a random graph from NetworkX library.*

````python
#Solution Milestone 1 question a) 

import networkx as nx
import random

def undirected_random_milestone1(num_nodes, p):
    """
    Constructs an undirected random graph using the Gilbert model.
    
    Parameters:
    - num_nodes (int): The number of nodes in the graph.
    - p (float): The probability of an edge existing between a pair of nodes.
    
    Returns:
    - G (networkx.Graph): The generated undirected random graph.
    """
    # Create an empty undirected graph
    G = nx.Graph()
    
    # Add nodes
    G.add_nodes_from(range(num_nodes))
    
    # Iterate over each pair of nodes and add an edge with probability p
    for i in range(num_nodes):
        for j in range(i + 1, num_nodes):  # Ensure we only consider each pair once
            if random.random() < p:  # Probability check
                G.add_edge(i, j)
    
    return G

````

````python
import matplotlib.pyplot as plt

# Generate the graph
G_gilbert = undirected_random_milestone1(10, 0.4)

# Define layout for the graph
pos = nx.spring_layout(G_gilbert)

# Plot the graph
nx.draw_networkx(G_gilbert, pos)
plt.title("Undirected Random Graph (Gilbert Model)")
plt.show()
````

b) Write the code to construct from scratch a **directed** random graph with 10 nodes using the Gilbert model, i.e., the probability of an edge existing between a pair of nodes is p. For this exercise use p = 0.4.

- *Note that you cannot use any of the functions that automatically creates a random graph from NetworkX library.*

````python
#Solution Milestone 1 question b)

import networkx as nx
import random

def directed_milestone1(num_nodes, p):
    """
    Constructs a directed random graph using the Gilbert model.
    
    Parameters:
    - num_nodes (int): The number of nodes in the graph.
    - p (float): The probability of an edge existing between a pair of nodes.
    
    Returns:
    - G (networkx.DiGraph): The generated directed random graph.
    """
    # Create an empty directed graph
    G = nx.DiGraph()
    
    # Add nodes
    G.add_nodes_from(range(num_nodes))
    
    # Iterate over each pair of nodes and add a directed edge with probability p
    for i in range(num_nodes):
        for j in range(num_nodes):  # Check all pairs including (i, i) for self-loops
            if i != j and random.random() < p:  # Probability check, excluding self-loops
                G.add_edge(i, j)
    
    return G

````

````python
import matplotlib.pyplot as plt

# Generate the directed graph
G_directed = directed_milestone1(10, 0.4)

# Define layout for the graph
pos = nx.spring_layout(G_directed)

# Plot the graph with directed edges
nx.draw_networkx(G_directed, pos, arrows=True)
plt.title("Directed Random Graph (Gilbert Model)")
plt.show()
````

````python
#Solution Milestone 2, question c) 

import networkx as nx
# Solution Milestone 2, question c)

# Closeness centrality = [(reachable nodes - 1) / sum of shortest path distances] * [(reachable nodes - 1)/(total number of nodes - 1)]

G_dir = directed_milestone1(10, 0.4)

def custom_closeness_centrality(G):
    num_nodes = G.number_of_nodes()
    custom_closeness_centrality = {} # Initialize dictionary that will store closeness centrality values
    n = len(G.nodes())  # Total number of nodes in the graph
    if G.is_directed():
        G = G.reverse()

    for node in G.nodes():
      # Calculate shortest paths from the node to all other nodes
      shortest_paths = nx.single_source_shortest_path_length(G, node)
      reachable_nodes = len(shortest_paths)  # Count the number of reachable nodes
      total_distance = sum(shortest_paths.values())  # Sum of shortest path distances

      # Calculate closeness centrality with NetworkX's normalization
      if total_distance > 0 and reachable_nodes > 1:
          # Apply the normalization factor to consider only reachable nodes
          custom_closeness_centrality[node] = ((reachable_nodes - 1) / total_distance) * ((reachable_nodes - 1) / (n - 1))
      else:
          custom_closeness_centrality[node] = 0.0  # Isolated node case
      
    return custom_closeness_centrality


closeness_centrality = custom_closeness_centrality(G_dir)
nx_closeness_centrality = nx.closeness_centrality(G_dir)

# Print and compare the results
print("Custom Closeness Centrality:", closeness_centrality)
print("NetworkX Closeness Centrality:", nx_closeness_centrality)
````

````python
#Solution Milestone 2, question e) 

import networkx as nx

# Step 1: Generate the directed random graph
G_directed = directed_milestone1(10, 0.4)

# Step 2: Custom computation of PageRank
def custom_pagerank(G, damping=0.85, tol=1e-5):
    num_nodes = G.number_of_nodes()
    # Initialize each node with equal PageRank value
    pagerank = {node: 1 / num_nodes for node in G.nodes()}
    new_pagerank = pagerank.copy()

    while True:
        for node in G.nodes():
            rank_sum = sum(pagerank[neighbor] / G.out_degree(neighbor)
                           for neighbor in G.predecessors(node))
            new_pagerank[node] = (1 - damping) / num_nodes + damping * rank_sum

        # Check convergence by comparing total change in PageRank values
        diff = sum(abs(new_pagerank[node] - pagerank[node]) for node in G.nodes())
        if diff < tol:
            break

        # Update pagerank for next iteration
        pagerank = new_pagerank.copy()

    return pagerank

# Calculate custom PageRank values
custom_pagerank_values = custom_pagerank(G_directed)

# Step 3: Compute PageRank using NetworkX for verification
nx_pagerank = nx.pagerank(G_directed, alpha=0.85)

# Step 4: Print custom results and compare with NetworkX results
print("Custom PageRank:", custom_pagerank_values)
print("NetworkX PageRank:", nx_pagerank)
````

````python
import networkx as nx

# Generate the undirected random graph
G_undirected = undirected_random_milestone1(10, 0.4)

# Custom computation of k-core decomposition with onion layers
def custom_k_core_decomposition_with_layers(G):
    core_number = {}
    onion_layers = {}
    G_copy = G.copy()
    current_core = 0
    current_layer = 0

    while G_copy.nodes():
        # Identify nodes with degree <= current_core
        nodes_to_remove = [node for node, degree in G_copy.degree() if degree <= current_core]
        
        if not nodes_to_remove:
            current_core += 1
            continue
            
        current_layer += 1
        
        # Assign core number and onion layer for removable nodes
        for node in nodes_to_remove:
            core_number[node] = current_core
            onion_layers[node] = current_core
            G_copy.remove_node(node)
        
    return core_number, onion_layers

# Calculate custom core number and onion layers
custom_core_number, custom_onion_layers = custom_k_core_decomposition_with_layers(G_undirected)

# NetworkX k-core decomposition for verification
nx_core_number = nx.core_number(G_undirected)
# Approximating NetworkX onion layers by sorting nodes by core number
nx_onion_layers = {node: nx_core_number[node] for node in nx.onion_layers(G_undirected)}

# Print results for comparison
# Print sorted results for comparison
print("Custom Core Number (sorted):  ", dict(sorted(custom_core_number.items())))
print("NetworkX Core Number (sorted):", dict(sorted(nx_core_number.items())))
print("Core Number Match:", custom_core_number == nx_core_number)

print("\nCustom Onion Layers (sorted):  ", dict(sorted(custom_onion_layers.items())))
print("NetworkX Onion Layers (sorted):", dict(sorted(nx_onion_layers.items())))
print("Onion Layers Match:", custom_onion_layers == nx_onion_layers)

````

````python
# Solution Milestone 4

import networkx as nx
import time
from networkx.algorithms.community import modularity, label_propagation_communities, greedy_modularity_communities
from community import community_louvain

# Function to compute modularity for a graph and list of communities
def custom_modularity(graph, communities):
    # Create a mapping from node to its community index for quick lookups
    node_to_community = {}
    for community_index, community in enumerate(communities):
        for node in community:
            node_to_community[node] = community_index

    # Total sum of all edge weights
    m = sum(graph[u][v].get('weight', 1) for u, v in graph.edges())
    modularity = 0.0

    # Iterate over all node pairs, but only compute contributions for pairs in the same community
    for i in graph.nodes():
        for j in graph.nodes():
            if node_to_community.get(i) == node_to_community.get(j):  # Only consider nodes in the same community
                # Adjacency matrix value (weighted or unweighted)
                A_ij = graph[i][j].get('weight', 1) if graph.has_edge(i, j) else 0
                # Weighted degree of nodes
                k_i = sum(graph[i][nbr].get('weight', 1) for nbr in graph.neighbors(i))
                k_j = sum(graph[j][nbr].get('weight', 1) for nbr in graph.neighbors(j))
                # Compute modularity contribution for the node pair
                modularity += A_ij - ((k_i * k_j) / (2 * m))

    # Final modularity value
    modularity /= (2 * m)
    return modularity

# Predefined graphs in NetworkX
graphs = {
    "Karate Club": nx.karate_club_graph(),
    "Davis Southern Women": nx.davis_southern_women_graph(),
    "Florentine Families": nx.florentine_families_graph(),
    "Les Miserables": nx.les_miserables_graph()
}

# Store results for modularity and execution time
results = {}

for graph_name, G in graphs.items():
    # print(f"\nProcessing {graph_name} graph...")
    results[graph_name] = {}

    # Label Propagation
    start_time = time.time()
    communities_lp = list(label_propagation_communities(G))
    end_time = time.time()
    custom_modularity_lp = custom_modularity(G, communities_lp)
    true_modularity_lp = modularity(G, communities_lp)
    results[graph_name]["Label Propagation"] = {
        "Custom Modularity": custom_modularity_lp,
        "True Modularity": true_modularity_lp,
        "Execution Time": end_time - start_time
    }

    # Greedy Modularity
    start_time = time.time()
    communities_gm = list(greedy_modularity_communities(G))
    end_time = time.time()
    custom_modularity_gm = custom_modularity(G, communities_gm)
    true_modularity_gm = modularity(G, communities_gm)
    results[graph_name]["Greedy Modularity"] = {
        "Custom Modularity": custom_modularity_gm,
        "True Modularity": true_modularity_gm,
        "Execution Time": end_time - start_time
    }

    # Louvain Method
    start_time = time.time()
    partition = community_louvain.best_partition(G)
    communities_louvain = {}
    for node, comm in partition.items():
        communities_louvain.setdefault(comm, []).append(node)
    end_time = time.time()
    custom_modularity_louvain = custom_modularity(G, communities_louvain.values())
    true_modularity_louvain = modularity(G, communities_louvain.values())
    results[graph_name]["Louvain"] = {
        "Custom Modularity": custom_modularity_louvain,
        "True Modularity": true_modularity_louvain,
        "Execution Time": end_time - start_time
    }

    # # Print results for each algorithm
    # for method, metrics in results[graph_name].items():
    #     print(f"{method} - Custom Modularity: {metrics['Custom Modularity']:.4f}, True Modularity: {metrics['True Modularity']:.4f}, Execution Time: {metrics['Execution Time']:.4f} seconds")

# Determine the best algorithm based on modularity and execution time
print("\nBest Algorithm Analysis:")
for graph_name, methods in results.items():
    print(f"\n{graph_name}:\n")
    for method, metrics in methods.items():
        print(f"Method:  {method}")
        print(f"Custom Modularity:  {metrics['Custom Modularity']}")
        print(f"NX Modularity:  {metrics['True Modularity']}")
        print(f"Execution Time:  {metrics['Execution Time']}\n")
    
    # Find the best modularity and best execution time
    best_modularity = max(methods, key=lambda x: methods[x]["True Modularity"])
    best_time = min(methods, key=lambda x: methods[x]["Execution Time"])
    print(f"Best Modularity: {best_modularity}")
    print(f"Best Execution Time: {best_time}\n")


````
